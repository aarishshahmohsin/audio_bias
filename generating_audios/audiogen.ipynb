```python
import torchaudio
from audiocraft.models import AudioGen
from audiocraft.data.audio import audio_write
```

```python
import pandas as pd
```

```python
model = AudioGen.get_pretrained('facebook/audiogen-medium', device='cuda')
```

```python
model.set_generation_params(duration=5)  
```

```python
template = "a {} talking"
```

```python
df = pd.read_csv("/kaggle/input/new-data/Bias_Types_Term_Category.csv")
```

```python
def generate_and_save(term, start_idx=0, number=100):
  descriptions = [template.format(term) for _ in range(number)]
  wav = model.generate(descriptions)

  for idx, one_wav in enumerate(wav):
    audio_write(f'./audio_lm/{term}_{idx+start_idx}', one_wav.cpu(), model.sample_rate, strategy="loudness", loudness_compressor=True)
```

```python
for idx, rows in df.iterrows():
  if rows['Category'] != 'Religion' and rows['Category'] != 'Portrayal in Media':
    print(idx)
    generate_and_save(rows['Term'], 1, 100)
```
